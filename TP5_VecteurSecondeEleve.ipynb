**3°) Des lunes**
* Exécuter le programme ci-dessous.

```python
from math import *
from random import *
import matplotlib.pyplot as plt


def distance(xA,yA,xB,yB):
    d=sqrt((xB-xA)**2+ (yB-yA)**2)   
    return d

xA=3.5
yA=5
xB=5
yB=5

plt.plot([xA,xB],[yA,yB],'b.')
plt.grid()
plt.show()
plt.axis("equal")

for i in range(1,5000):
    x1=random()*10
    y1=random()*10
    dA=distance(x1,y1,xA,yA)
    dB=distance(x1,y1,xB,yB)
    if dA>3 and dB<4:
        plt.plot(x1,y1,'r.')
```
